```python
def deduplication(x):
    trim = x.strip()
    index = (trim+trim).find(trim, 1)
    if index != -1:
        return trim[:index]
    else:
        return trim
```

```python
df_admin['location'] = df_admin['location'].apply(deduplication)
df_admin['location']
```

```python
# Fill any NaN value in the column of area with unknown
# Be aware of the parameter inplace, if true, will modify any other views on this object.
df_admin['area'].fillna(value='unknown', inplace = True)
df_admin['area'] = df_admin['area'].apply(deduplication)
df_admin['area']
```

```python
def apply_posted_time(x):
    if 'd' in x:
        return -int(x[:-1])
    if 'm' in x:
        return -(int(x[:-1])*30)
    else:
        return 0
```

```python
df_admin['posted time'] = df_admin['posted time'].apply(apply_posted_time)
df_admin
```
